### 13 Number Purchases in AM and PM

```python
df['AM or PM'].value_counts()
```

```python
df['AM or PM'].value_counts().plot(kind='bar', figsize=(5,4))
plt.title("Purchases During AM vs PM")
plt.xlabel("Time of Day")
plt.ylabel("Number of Purchases")
plt.show()
```

**Insight: Shows whether customers prefer shopping during AM or PM hours.**

### 14. Credit Cards Expiring in 2020

```python
df[df['CC Exp Date'].str.contains('20')].shape[0]
```

```python
def fun():
    count=0
    for data in df['CC Exp Date'] :
           if data.split('/')[1]=='20':
            count = count+1
    print(count)
```

```python
fun()
```

**Insight: Identifies users whose credit cards expire in 2020, useful for renewal or reminder campaigns.**

### 15.Top 5 most popular email providers (e.g. gmail.com, yahoo.com, etc...) 

```python
df['temp']= df['Email'].str.split("@").str[1]
df['temp'].value_counts().head()
```

```python
df['temp'].value_counts().head(5).plot(kind='bar', figsize=(6,4))
plt.title("Top 5 Email Providers")
plt.xlabel("Email Provider")
plt.ylabel("Number of Users")
plt.show()
```

### 16.Top 5 Job Titles by Average Purchase Price

```python
df.groupby('Job')['Purchase Price'].mean().sort_values(ascending=False).head(5)
```
