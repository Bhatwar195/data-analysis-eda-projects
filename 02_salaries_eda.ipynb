### 8.Top 5 Employee Name Occurrences

```python
df['EmployeeName'].value_counts().head(5).plot(kind='barh', figsize=(6,4))
plt.title("Top 5 Employee Name Occurrences")
plt.ylabel("Employee Name")
plt.xlabel("Count")
plt.show()
```

### 11.BasePay Summary Statistics(Find Minimum, Maximum, and Average BasePay)

### 12.Replace 'Not Provided' with NaN in Employee Name

```python
df['EmployeeName']=df['EmployeeName'].replace('Not Provided',np.nan)
```

Standardizes missing values by converting placeholder text into proper NaN values for accurate analysis.

### 13. Display All the Employee Names From Fire Department

```python
df[df['JobTitle'].str.contains('FIRE',case=False)]['EmployeeName']
```

Lists all employees associated with the Fire Department, useful for department-level analysis or audits.

### 14.Drop the rows having more than five missing value in a row 

```python
df.drop(df[df.isnull().sum(axis=1)>5].index,axis=0)
```

Removes records with excessive missing data to improve data reliability and analysis accuracy.`

### 18. Find Average BasePay of All Employee Per Year

```python
avg_year=df.groupby('Year')['BasePay'].mean()
```

```python
avg_year.plot(kind='line', figsize=(6,4))
plt.title("Average BasePay per Year")
plt.xlabel("Year")
plt.ylabel("Average BasePay")
plt.show()
```

Shows how the average base pay changes across different years, helping identify salary trends over time.

### 19.Find Average BasePay of Employee Having Job Title ACCOUNTANT

```python
df[df['JobTitle'].str.contains('ACCOUNTANT',case=False)]['BasePay'].mean()
```

Displays the typical base pay for employees with the Accountant job title.

### 20.Find Top 5 Most Common Jobs

```python
df['JobTitle'].value_counts().head(5).plot(kind='barh')
plt.title("Top 5 Most Common Job Titles")
plt.ylabel("Job Title")
plt.xlabel("Count")
plt.show()
```

### 21.Displays the typical base pay for employees with the Accountant job title.

```python
df.groupby('JobTitle')['BasePay'].mean()
```

```python
top_avg_jobs = (
    df.groupby('JobTitle')['BasePay']
    .mean()
    .sort_values(ascending=False)
    .head(5)
)
top_avg_jobs
```

```python
import matplotlib.pyplot as plt

top_avg_jobs.plot(kind='barh', figsize=(6,4))
plt.title("Top 5 Job Titles by Average BasePay")
plt.ylabel("Job Title")
plt.xlabel("Average BasePay")
plt.show()
```
